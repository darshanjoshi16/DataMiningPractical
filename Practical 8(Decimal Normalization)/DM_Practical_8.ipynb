```python
#implementation of decimal score normalization

#importing the library
import pandas as pd

#defining the function which will be doing the decimal normalization
def decNorm(num,maxNum):
    #it gives the value of j which is used in getting the value of 10 raised to j by taking the maxlength of the element in the list and converting into string
    digit=int(len(str(maxNum)))

    #computing 10 raised to j
    div=pow(10,digit)

    #returning the result
    return num/div

#driver code for the dynamic implementation of our function
print("==========DECIMAL NORMALIZATION========")
n = int(input("For How Many records do you want to calculate Decimal normalization?: "))

values = [int(x) for x in input().split(',')] 
while(True):
            print("==========Choices====================")
            print("\n 1: Normalization of single data value")
            print("\n 2: Normalization of whole dataset")
            print("\n====================================")
            choice = int(input("Please Enter the choice: "))
            print("\n====================================")

            if choice == 1:
              a = int(input("Enter the value you want to normalize:"))
              print("\n==============================================")
              if(a in values):
                result = decNorm(a,max(values))
                print("Entered  "+"Normalized")
                print(str(a)+"       "+str(result))
              else:
                print("Entered value is not available in list.Enter Correct value")

            if choice == 2:
              result = []

              for i in values:
                temp = decNorm(i,max(values))
                result.append(temp)

              df = pd.DataFrame(result,values)
              df.columns = ['Normalized Values']
              print(df)
```
