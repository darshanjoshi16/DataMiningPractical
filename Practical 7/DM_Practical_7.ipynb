```python
#Approach 2: Using SciPy Library

#importing the library for completion of task
import numpy as np
import pandas as pd
import scipy.stats as stats

#taking the number of records and data of records from user
#store the data into list
n = int(input("For How Many records do you want to calculate z-score normalization?: "))
values = [float(x) for x in input().split(',')] 

#converting list into numpy array object
val_array = np.array(values)

#using Scipy.stats function's zscore calculate the z-score
normalized_value = stats.zscore(val_array,axis=0,ddof=0)

#creating the dataframe and displaying the result
dataf_1 = pd.DataFrame(normalized_value,val_array)
dataf_1.columns = ['Normalized Values']
print(dataf_1)

```

```python
#Approach 1: Using Conventional Methods

#importing the libraries
import pandas as pd

#defining the Z-score function as per our requirement
def z_score(values):

  #taking the mean of list generated by user input
  mean = sum(values) / len(values)
  
  #calculating the difference
  differences = [(value - mean)**2 for value in values]
  
  #calculating the sum of difference and S.D
  sum_of_differences = sum(differences)
  standard_deviation = (sum_of_differences / (len(values) - 1)) ** 0.5
  
  #calculating Z-score values for each value from input
  zscores = [(value - mean) / standard_deviation for value in values]
  return zscores
             

#driver code for implementation
n = int(input("For How Many records do you want to calculate z-score normalization?: "))

values = [float(x) for x in input().split(',')] 
result = z_score(values)

dataf_1 = pd.DataFrame(result,values)
dataf_1.columns = ['Normalized Values']
print(dataf_1)

```
