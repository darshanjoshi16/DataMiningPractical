```python
#Importing the libraries

import numpy as np
import pandas as pd
```

```python
# Importing the dataset
df = pd.read_csv("https://raw.githubusercontent.com/nyuvis/datasets/master/auto/imports-85.data")
```

```python
#Columns with missing values
df.apply(lambda row:row[row=='?'].index,axis=0)
```

```python
#Data Preprocessing methods
df["normalized-losses"].replace("?",120, inplace= True)
df["num-of-doors"].replace("?","two", inplace= True)
df["bore"].replace("?",3.00, inplace= True)
df["stroke"].replace("?",3.00, inplace= True)
df["horsepower"].replace("?",120, inplace= True)
df["peak-rpm"].replace("?",5000, inplace= True)
df["price"].replace("?",12000, inplace= True)
```

```python
#checking after this pre-processing
df.apply(lambda row:row[row=='?'].index,axis=0)
```
